```python
# Property 1: Set does't support indexing
set_var[0]
```

```python
# Property 2: Set does't have any order
set_var={4,'P','Data Science',87,14.5}
print(set_var)
```

```python
set_var={4,87,1,98,2,45,3}
print(set_var)
```

```python
# Property 3: Set can  be changed/ mutable
set_var={4,'P','Data Science',87,14.5}
print(set_var)
set_var.add(999)
set_var
```

```python
set_var={4,'P','Data Science',87,14.5}
print(set_var)
set_var.add(999)
set_var
```

```python
set_var={4,'P','Data Science',87,14.5}
print("Old set values",set_var)
set_var.discard(99)
print("New set values",set_var)
```
