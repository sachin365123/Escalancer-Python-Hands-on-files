# Bar chart

```python
y=[3,10,7,1,6,8]
x=range(len(y))
width=1/2
plt.bar(x,y,width,color='red')
plt.xlabel('Current-->')
plt.ylabel('Velocity-->')
plt.title("Bar chart")
plt.show()
```

# Bubble Graph

```python
N=50
x=np.random.rand(N)
y=np.random.rand(N)
colors=np.random.rand(N)
area=np.pi*(15*np.random.rand(N))**2
plt.scatter(x,y,s=area,c=colors,alpha=0.8)
plt.show()
```

# Pie chart 

```python
l='Data Science','Machine Learning','Power BI','SQL'
sizes=[15,30,45,10]
kapil=(0.2,0,0,0)
fig1,ax1=plt.subplots()
ax1.pie(sizes,explode=kapil,labels=l,autopct='%1.6f%%',shadow=True,startangle=180)
ax1.axis('equal')
plt.show()
```

# Subplot

```python
import matplotlib.pyplot as plt

y = [3, 10, 7, 1, 6, 8]
x = range(len(y))
width = 1/2
colors = ['red', 'blue', 'green', 'orange', 'purple', 'yellow']  # Specify colors for each bar

fig, axs = plt.subplots(2, 2)  # Create a 2x2 grid of subplots

# Plot each bar chart in a subplot
for i, ax in enumerate(axs.flat):
    if i < len(y):
        ax.bar(x, y, width=width, color=colors[i])
        ax.set_xlabel('Current-->')
        ax.set_ylabel('Velocity-->')
        ax.set_title(f"Bar chart {i+1}")

plt.tight_layout()  # Adjust layout to prevent overlap
plt.show()
```
