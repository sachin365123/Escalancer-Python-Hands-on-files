Import basic libraries

```python
import numpy as np
from keras.models import Sequential
from keras.layers import Dense
```

Create input and output records

```python
# AND gate truth table
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
Y = np.array([[0], [0], [0], [1]])
```

# Create a sequential model

```python
model=Sequential()
```

# Add a dense layer with one neuron and sigmoid activation function

```python
model.add(Dense(units=1, input_dim=2, activation='sigmoid'))
```

# Compile the model

```python
model.compile(loss='mean_squared_error', optimizer='adam', metrics=['accuracy'])
```

# Train the model

verbose=0: No output will be displayed during the training process.
verbose=1: Displays a progress bar for each epoch, showing the training progress.
verbose=2: Displays one line per epoch, providing a summary of the training progress.

```python
model.fit(X, Y, epochs=1000, verbose=2)
```

# Test the model

```python
print("Predictions:")
print(model.predict(X))
```

```python
print("Predictions:")
print(model.predict(X))
```

# Evaluate the model

```python
loss, accuracy = model.evaluate(X, Y)
print(f"Loss: {loss:.4f}, Accuracy: {accuracy * 100:.2f}%")
```

XOR table

```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import SGD

# XOR gate truth table
data = np.array([[0, 0],
                 [0, 1],
                 [1, 0],
                 [1, 1]])

labels = np.array([0, 1, 1, 0])
```

```python
# Create a sequential neural network model
model = Sequential()

# Add a hidden layer with 2 neurons and input layer with 2 neurons
model.add(Dense(2, input_dim=2, activation='sigmoid'))
```

```python
# Add an output layer with 1 neuron
model.add(Dense(1, activation='sigmoid'))
```

```python
# Compile the model
sgd = SGD(learning_rate=0.1)
model.compile(loss='mean_squared_error', optimizer=sgd)
```

```python
# Train the model
epochs = 100
model.fit(data, labels, epochs=epochs, verbose=1)
```

```python
# Evaluate the model
print("Training completed. Evaluating the model...\n")

for i in range(len(data)):
    input_data = data[i].reshape(1, 2)
    predicted_label = model.predict(input_data)[0][0].round()
    print(f"Input: {data[i]}, Predicted Output: {predicted_label}")
```

import library

```python
import pandas as pd
import tensorflow as tf
from sklearn.model_selection import train_test_split
```

# Load the dataset from a CSV file

```python
data = {
    'Travel': [1.2, 2.1, 3.5, 4.0, 5.8],
    'Manufucturing': [0.5, 1.1, 2.6, 3.2, 4.9],
    'Proft_or_Loss': [0, 1, 0, 1, 0]
}
```

```python
df=pd.DataFrame(data)
```

```python
df
```

```python
type(df)
```

# Split the dataset into training and testing sets

```python
train_df, test_df = train_test_split(df, test_size=0.2)
```

```python
train_df
```

```python
test_df
```

# Prepare the input features and target labels

```python
train_features = train_df[['Travel', 'Manufucturing']].values
train_labels = train_df['Proft_or_Loss'].values
test_features = test_df[['Travel', 'Manufucturing']].values
test_labels = test_df['Proft_or_Loss'].values
```

# Create the neural network model

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(16, activation='relu', input_shape=(2,)),
    tf.keras.layers.Dense(1, activation='sigmoid')
])
```

# Compile the model

```python
model.compile(optimizer='adam',loss='binary_crossentropy', metrics=['accuracy'])
```

# Train the model

```python
model.fit(train_features, train_labels, epochs=10, batch_size=1, verbose=1)
```

```python
model.fit(train_features, train_labels, epochs=10, batch_size=2, verbose=1)
```

# Evaluate the model on the testing set

```python
loss, accuracy = model.evaluate(test_features, test_labels, verbose=0)
print(f'Test loss: {loss:.4f}')
print(f'Test accuracy: {accuracy:.4f}')
```
